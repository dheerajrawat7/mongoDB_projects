```python
import pandas as pd

client = MongoClient('mongodb://localhost:27017/')
db = client['covid_db']  
collection = db['covid_db_coll']

df = pd.read_csv("worldwide_covid_sample_data(1).csv")

data = df.to_dict(orient='records')

collection.insert_many(data)

print("CSV file inserted into MongoDB successfully")
print(df)
```

1. Write a query to retrieve the population which are greater than 3711 and belongs to the
#Colombia administrative area, sorted by their hosp in descending order and also count
#the total resulted documents.

```python
result = collection.find({"population":{"$gt":3711},"administrative_area_level_1":"Colombia"},{"hosp":1,"population":1,"administrative_area_level_1":1,"_id":0}).sort("hosp",-1)
for i in result:
    print(i)
```

2. Write a query to find all administrative areas where confirmed is greater than 4534 and
also count the total resulted documents

```python
result = collection.find({"confirmed": {"$gt": 4534}}, {"administrative_area_level_1": 1, "_id": 0})
count = collection.count_documents({"confirmed": {"$gt": 4534}})
print("Total documents:", count)
for i in result:
    print(i)
```

3. Write a query to find all administrative areas where the testing policy is either 2 or 3.

```python
result = collection.find({"testing_policy": {"$in": [2, 3]}}, {"administrative_area_level_1": 1, "_id": 0})
count = collection.count_documents({"testing_policy": {"$in": [2, 3]}})
print("Total documents:", count)
for i in result:
    print(i)
```

4. Write a query to calculate the total population of all the countries.

```python
pipeline = [
    {"$group": {"_id": None, "total_population": {"$sum": "$population"}}}
]

result = collection.aggregate(pipeline)

for doc in result:
    print("Total population of all countries:", doc["total_population"])
```

8. Write a query to calculate the total count of population within each administrative area
level 1 group.

```python
pipeline = [
    {"$group": {"_id": "$administrative_area_level_1", "total_population": {"$sum": 1}}}
]

result = collection.aggregate(pipeline)

for doc in result:
    print("Administrative Area Level 1:", doc["_id"])
    print("Total Population Count:", doc["total_population"])
```

9. Write a query to calculate the total sum of population with each administrative area
level 1 group.

```python
pipeline = [
    {"$group": {"_id": "$administrative_area_level_1", "total_population": {"$sum": "$population"}}}
]

result = collection.aggregate(pipeline)

for doc in result:
    print("Administrative Area Level 1:", doc["_id"])
    print("Total Population:", doc["total_population"])
```

10.  Write a query to find the minimum and maximum population with each administrative
area level 1 group.

```python
pipeline = [
    {"$group": {
        "_id": "$administrative_area_level_1",
        "min_population": {"$min": "$population"},
        "max_population": {"$max": "$population"}
    }}
]

result = collection.aggregate(pipeline)

for doc in result:
    print("Administrative Area Level 1:", doc["_id"])
    print("Minimum Population:", doc["min_population"])
    print("Maximum Population:", doc["max_population"])
```

11. Write a query to grouping the id by administrative area level 1 and administrative area
level 2 and find the average population.

```python
pipeline = [
    {"$group": {
        "_id": {
            "admin_area_level_1": "$administrative_area_level_1",
            "admin_area_level_2": "$administrative_area_level_2"
        },
        "average_population": {"$avg": "$population"}
    }}
]

result = collection.aggregate(pipeline)

for doc in result:
    print("Administrative Area Level 1:", doc["_id"]["admin_area_level_1"])
    print("Administrative Area Level 2:", doc["_id"]["admin_area_level_2"])
    print("Average Population:", doc["average_population"])
```
